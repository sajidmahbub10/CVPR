```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(-10, 11, 1)
print(x)
```

```python
def plot_graph(y,ylabel):
    plt.figure()
    plt.plot(x,y, 'o--')
    plt.grid(True)
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()
```

# Step Activation Function

```python
y = np.zeros(len(x))
y[x>=0] = 1

plot_graph(y, "Step(x)")
```

# Sigmoid Activation Function

```python

y = 1/(1 + np.exp(-x))

plot_graph(y, "Sigmoid(x)")
```

# TanH Activation Fuction

```python
y = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
plot_graph(y, "TanH(X)")
```

# ReLU Activation Function

```python
y = np.maximum(0, x)

plot_graph(y, "ReLU(x)")
```

# ELU Activation Function

```python
alpha = 1.0

y = np.where(x>=0, x, alpha*(np.exp(x)-1))

plot_graph(y, "ELU(x)")
```

# SELU Activation Function

```python
 def selu(x, scale=1.0507, alpha=1.6733):
    return np.where(x > 0, scale*x, scale*(alpha*np.exp(x)-alpha))

plot_graph(selu(x), "SELU(x)")
```
